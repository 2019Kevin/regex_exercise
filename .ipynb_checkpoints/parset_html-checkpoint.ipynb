# <strong>解析拉勾网一个招聘页面的信息掌握正则表达式的基本操作.</strong><br>

```python
import requests
import re
from pprint import pprint

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/60.0.3112.113 Safari/537.36",
}


wb_data = requests.get(url="https://www.lagou.com/jobs/3584592.html", headers=headers)
html = wb_data.text
```

# <strong>解析页面信息</strong>

```python
company_info = re.findall('<dl class="job_company" id="job_company">[\s\S]*?</dl>', html)[0]
work_addr = re.findall('<div class="work_addr">[\s\S]*?</div>', html)[0]
job_title = re.findall('<ul class="position-label clearfix">[\s\S]*?</ul>', html)[0]
job_request = re.findall('<dd class="job_request">[\s]*?<p>[\s\S]*?</p>', html)[0]
job_description = re.findall('<h3 class="description">职位描述：</h3>[\s\S]*?<div>[\s\S]*<p>[\s\S]*?<br>', html)[0]
job_task = job_description.split(r'<p>工作内容：</p>')[0]
job_necessary = job_description.split(r'<p>工作内容：</p>')[1]



job_name = re.findall('<div class="job-name" title=(.*?)>', html)[0]  
company_name = re.findall('<img class="b2" .*? alt=(.*?)>', html)[0].replace("/", "").strip()
company_website = re.findall('<a .*?>(.*?)</a>\s+?<span class="hovertips">公司主页</span>', html)[0].strip()
field = re.findall('</i>(.*?)\s+?<span class="hovertips">(.*?)</span>', company_info)[0][0]
company_stage = re.findall('</i>(.*?)\s+?<span class="hovertips">(.*?)</span>', company_info)[1][0]
company_size = re.findall('</i>(.*?)\s+?<span class="hovertips">(.*?)</span>', company_info)[3][0]
work_addr = "-".join(re.findall('<a .*?>(.*?)</a>', work_addr)[:-1] + re.findall('\s+?- (.*?)\n[\s\S]*?', work_addr))
job_title = '/'.join(re.findall('<li class="labels">(.*?)</li>', job_title))
salary = re.findall('<span[\s\S]*?>(.*?)</span>', job_request)[0].strip()
experience = re.findall('<span[\s\S]*?>(.*?)</span>', job_request)[2].replace("/", "").replace("经验", "")
degress =  re.findall('<span[\s\S]*?>(.*?)</span>', job_request)[3].replace("/", "").strip()
job_task = re.findall('<p>(.*?)</p>', job_task)[1:-1]
job_necessary = re.findall('<p>(.*?)</p>', job_necessary)
```

```python
position_dict = {
    "职位名称": job_name,
    "公司名称": company_name,
    "公司网站": company_website,
    "所处领域": field,
    "发展阶段": company_stage,
    "公司规模": company_size,
    "工作地址": work_addr,
    "职位标签": job_title,
    "薪资水平": salary,
    "经验要求": experience,
    "学历要求": degress,
    "工作任务": job_task,
    "工作要求": job_necessary,
}

pprint(position_dict)
```

# <strong>将工作任务和工作要求写入到txt文件中</strong><br>

```python
with open("job_task.txt", "w", encoding="utf-8") as f:
    for item in job_task:
        f.write(item + "\n")

with open("job_necessary.text", "w", encoding="utf-8") as f:
    for item in job_necessary:
        f.write(item + "\n")
```

<strong>使用python绘制词云</strong>

```python
import matplotlib.pyplot as plt              
from wordcloud import WordCloud,STOPWORDS                                 # 词云生成库
import jieba                                                    # jieba分词库
%matplotlib inline
```

# <strong>职位要求</strong>词云图

```python
with open("job_necessary.text", "r", encoding="utf-8") as f:
    text = f.read()
    
    wordList_after_jieba = jieba.cut(text, cut_all=True)        # 使用jieba分词 
    wl_space_split = " ".join(wordList_after_jieba)             #  将分词的结果以空格分隔
    
    job_necessary_wordcloud = WordCloud(
        background_color = "white",                             # 设置背景颜色
#         mask = plt.imread("beautifulgirl.jpg"),                 # 设置背景图片
        max_words = 1000,                                       # 设置最多显示的字数
        stopwords = STOPWORDS,                                  # 设置停用词
        font_path="C:/Windows/Fonts/MSYH.TTC",                  # 设置字体
        random_state = 30,                                      # 设置随机状态
    ).generate(wl_space_split)                                  # 对分词后的文本产生词云
    
    fig, ax = plt.subplots()                                    # 开始绘图
    fig.set_size_inches(8, 4)
    plt.imshow(job_necessary_wordcloud)
    plt.show()
```

# <strong>职位内容</strong>词云图

```python
with open("job_task.txt", "r", encoding="utf-8") as f:
    text = f.read()
    
    wordList_after_jieba = jieba.cut(text, cut_all=True)        # 使用jieba分词 
    wl_space_split = " ".join(wordList_after_jieba)             #  将分词的结果以空格分隔
    
    job_necessary_wordcloud = WordCloud(
        background_color = "white",                             # 设置背景颜色
#         mask = plt.imread("beautifulgirl.jpg"),                 # 设置背景图片
        max_words = 1000,                                       # 设置最多显示的字数
        stopwords = STOPWORDS,                                  # 设置停用词
        font_path="C:/Windows/Fonts/MSYH.TTC",                  # 设置字体
        random_state = 30,                                      # 设置随机状态
    ).generate(wl_space_split)                                  # 对分词后的文本产生词云
    
    fig, ax = plt.subplots()                                    # 开始绘图
    fig.set_size_inches(8, 4)
    plt.imshow(job_necessary_wordcloud)
    plt.show()
```
